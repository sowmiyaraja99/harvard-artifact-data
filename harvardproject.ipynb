```python
url="https://api.harvardartmuseums.org/object"
pages=25
classifications=["Paintings"]
records1=[]
records2=[]
records3=[]
for classification in classifications:
   for page_num in range(1,pages+1):
        params={"apikey":api_key,"size":100,"page":page_num,"classification":classification}
        response=requests.get(url,params)
        data=response.json()


        for i in data['records']:
         records1.append(dict(id=i.get('id'),title=i.get('title'),culture=i.get('culture'),period=i.get('period'),
        century=i.get('century'),medium=i.get('medium'),dimensions=i.get('dimensions'),description=i.get('description'),
        department=i.get('department'),classification=i.get('classification'),
        accessionyear=i.get('accessionyear'),accessionmethod=i.get('accessionmethod')))

         records2.append(dict(
                objectid=i.get('objectid'),
                imagecount=i.get('imagecount'),
                mediacount=i.get('mediacount'),
                colorcount=i.get('colorcount'),
                rank=i.get('rank'),
                datebegin=i.get('datebegin'),
                dateend=i.get('dateend')))
         records3.append(dict(objectid=i.get('objectid'),
         color=json.dumps(i.get('colors',[])),
         spectrum=i.get('spectrum'),
         hue=i.get('hue'),
         percent=i.get('percent'),
         css3=i.get('css3')))
```
